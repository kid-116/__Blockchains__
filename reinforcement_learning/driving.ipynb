# Racing

```python
import gym
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy
import os
```

## Env

```python
ENV_NAME = 'CarRacing-v0'
```

#### Testing

```python
env = gym.make(ENV_NAME)

episodes = 2
for ep in range(1, episodes+1):
    obs = env.reset()
    done = False
    score = 0
    while not done:
        env.render()
        action = env.action_space.sample()
        obs, reward, done, info = env.step(action)
        score += reward
    
    print(f"Ep #{ep} - Score: {score}")
env.close()
```

### Vectorized Stack of Envs

```python
env = gym.make(ENV_NAME)
env = DummyVecEnv([lambda: env])
```

## Model

### Initializing

```python
LOG_PATH = os.path.join('training', 'logs', ENV_NAME)
```

```python
# new
model = PPO('CnnPolicy', env, verbose=1, tensorboard_log=LOG_PATH)
```

```python
# load
MODEL_PATH = os.path.join('training', 'saved_models', ENV_NAME, 'v1')
model = PPO.load(MODEL_PATH, env=env)
```

### Training

```python
model.learn(total_timesteps=100000)
```

```python
# save model
SAVE_PATH = os.path.join('training', 'saved_models', ENV_NAME, 'v1')
model.save(SAVE_PATH)
del model
```

### Evaluation

```python
evaluate_policy(model, env, n_eval_episodes=10, render=True)
# returns reward, std deviation
```

```python
env.close()
```

### Test

```python
episodes = 2
for ep in range(1, episodes+1):
    obs = env.reset()
    done = False
    score = 0
    while not done:
        env.render()
        action, _ = model.predict(obs)
        obs, reward, done, info = env.step(action)
        score += reward
    
    print(f"Ep #{ep} - Score: {score}")
env.close()
```
